## Data Cleaning

### Remove URLs, Hashtags, Mention & Retweet

```python
%%time 
import re

# remove hyperlinks
tsel["Clean Text"] = [re.sub(r'https?:\/\/.\S+', "", row) for row in tsel['Text']]
tsel["Clean Text"] = [re.sub(r'https?:\/\/.\S+', "", row) for row in tsel['Clean Text']]

# Remove Hashtags
tsel['Clean Text'] = [re.sub(r'#\S+', '', row) for row in tsel['Clean Text']]

# Remove RT
tsel['Clean Text'] = [re.sub(r'^RT[\s]+', '', row) for row in tsel['Clean Text']]

# Remove Mentions
tsel['Clean Text'] = [re.sub(r'@\S+', '', row) for row in tsel['Clean Text']]

```

### Remove Punctuation & Casefolding

```python
%%time
import string as st
# Converting all words to lower case and removing punctuation
tsel['Clean Text'] = [re.sub(r'\d+\S*', '',
                  row.lower().replace('.', ' ').replace('_', '').replace('/', ''))
                  for row in tsel['Clean Text']]
# Remove punctuation
tsel['Clean Text'] = [re.sub(r'[!\n"#$%&()*+,-./:;<=>?@[\]^_`{|}~]', ' ', row)
                  for row in tsel['Clean Text']]
# Remove numbers
tsel['Clean Text'] = [re.sub(r'\d+', '', row) for row in tsel['Clean Text']]

# Remove Emoji
emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  # emoticonsx
                u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                u"\U0001F680-\U0001F6FF"  # transport & map symbols
                u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                u"\U00002702-\U000027B0"
                u"\U000024C2-\U0001F251"
                u"\U0001f926-\U0001f937"
                u'\U00010000-\U0010ffff'
                u"\u200d"
                u"\u2640-\u2642"
                u"\u2600-\u2B55"
                u"\u23cf"
                u"\u23e9"
                u"\u231a"
                u"\u3030"
                u"\ufe0f"
    "]+", flags=re.UNICODE)

tsel['Clean Text'] = [emoji_pattern.sub(r'', row) for row in tsel['Clean Text']]

#Remove Whitespace
tsel['Clean Text'] = [re.sub(r'\s+', ' ', row).strip() for row in tsel['Clean Text']]

#Remove special character(bukan huruf)
tsel['Clean Text'] = [re.sub('[^a-zA-Z]', ' ', row) for row in tsel['Clean Text']]

tsel
```

### Tokenization

```python
%%time
import nltk
nltk.download("stopwords")

# Tokenizing comments and putting them into a new column
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')  # by blank space
tsel['tweet_tokens'] = tsel['Clean Text'].apply(tokenizer.tokenize)
tsel
```

### Convert Slang Words

```python
slang_dict = pd.read_csv("newslangword_dict.csv")
word = {}

for idx, row in slang_dict.iterrows():
    if row[0] not in word:
        word[row[0]] = row[1]
        
def ConvertSlangword(document):
    return [word[term] if term in word else term for term in document]

tsel['convSlangWord'] = tsel['tweet_tokens'].apply(ConvertSlangword)

tsel
```

### Remove Stopwords

```python
import nltk 
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# Mendapatkan stopword bahasa Indonesia dari NLTK stopword
stopwords_list = stopwords.words('indonesian')

# Menambahkan stopword tambahan
stopwords_list.extend(['pas', 'ya', 'sih', 'deh', 'loh', 'oiya', 'nih', 'ok', 'ah',
                       'yg', 'dg', 'dgn', 'ny', 'd', 'klo', 'kalo', 'amp', 'biar', 
                       'bikin', 'bilang', 'gak', 'ga', 'krn', 'nya', 'nih', 'sih',
                       'si', 'tau', 'tuh', 'utk', 'iya', 'jd', 'jgn', 'sdh',
                       'aja', 'n', 't', 'nyg', 'hehe', 'pen', 'u', 'nan', 'loh',
                       'mmm', 'yah', 'kak', 'ayo', 'banget', 'annya', 'gxh', 'di'
                       'by', 'ke', 'mah', 'ini', 'dongg', 'huaaa', 'hlaaaah', 'an'
                       'eeuyyy', 'bla', 'gass', 'kakak', 'bang', 'anak', 'tertawa',
                       'terima', 'kasih', 'semoga', 'orang', 'coba', 'amin','terimakasih',
                       'hmmmmmmmm', 'kah','an','a','wkakaka','x','lho','gih','yakk','miui',
                       'xx','kh','sich''hhh','tl','ta','va','ahjussi','yach','ef','ko','mai','bl','h'])

# Mengkonversi list ke dictionary
stopwords_list = set(stopwords_list)

# Menghapus stopword pada data
def remove_stop(words):
    return [word for word in words if word not in stopwords_list]

tsel['stopword'] = tsel['convSlangWord'].apply(remove_stop) 
print(stopwords_list)
tsel['stopword']
```

### Stemming

```python
# import Sastrawi package
!pip install Sastrawi
!pip install swifter
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter
```

```python
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in tsel['stopword']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)
print("------------------------")


# apply stemmed term to dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

tsel['text_result'] = tsel['stopword'].swifter.apply(get_stemmed_term)
# print(tsel['tokens_stemmed'])
```
