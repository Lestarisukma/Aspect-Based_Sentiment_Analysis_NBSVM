```python
# Library gensim
import gensim
from gensim.models import FastText
from gensim.models import TfidfModel
from gensim.corpora import Dictionary


import pandas as pd 
import numpy as np
import nltk
from tqdm.auto import tqdm

from nltk.tokenize import word_tokenize
from gensim.models import FastText

import os 
working_directory = os.getcwd()
print(working_directory)
```

# FastText Corpus Twitter

```python
df_twt = pd.read_csv(r"prepTwitter.csv", usecols=['Clean Text'])
```

```python
df_twt.rename(columns = {'Clean Text':'clean_text'}, inplace = True)
df_twt
```

```python
df_twt['clean_text'] = df_twt['clean_text'].apply(str)
df_twt['clean_text'] = df_twt['clean_text'].astype(np.uint8,errors='ignore')
```

```python
def tokenizing(data):
    return nltk.tokenize.word_tokenize(data)
```

```python
df_twt['clean_text'] = df_twt['clean_text'].apply(tokenizing)
df_twt
```

```python
model_twt= FastText(sentences = df_twt['clean_text'], window = 5, min_count = 3, workers = 4, sg = 0, hs = 0)
```

```python
model_twt.save('model_twitter')
```

#### Save Model FastText Twitter

```python
os.makedirs("model_twt/fasttext/", exist_ok=True)
```

```python
model_twt.save("model_twt/fasttext/prepTwitter.fasttext")
```

```python
model_twt = FastText.load("model_twt/fasttext/prepTwitter.fasttext")
```

#### Model Information

```python
ft_twt = model_twt.wv
```

```python
ft_twt.vectors
```

```python
# coba 
ft_twt["jlek"]
```

### Sanity Check

```python
# Train Similarity Model FastText
ft_twt.most_similar('admin')
```

```python
def word_similarity(text):
    x = []
    for i in text:
        x.append(i[0])
    return x

def create_corpus10(text):
    word = ft_twt.most_similar(text, topn=10)
    return word

words10 = ft_twt.key_to_index

corpus10 = list(map(create_corpus10, words10))
cft10 = pd.DataFrame(corpus10, index=words10)
cft10['All'] = corpus10
cft10['similarity'] = cft10['All'].apply(word_similarity)

cft10.to_excel('corpus_tweet.xlsx')
```

```python
cft10
```
