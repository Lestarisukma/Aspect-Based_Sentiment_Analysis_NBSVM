```python
import ast

def join_text_list(text):
    texts = ast.literal_eval(text)
    return ' '.join([text for text in texts])

tsel['text_result'] = tsel['text_result'].apply(join_text_list)
tsel
```

## Ektraksi Fitur TF-IDF 

```python
%%time
# Creating the sklearn object
tfidf = sklearn.feature_extraction.text.TfidfVectorizer(ngram_range=(1, 2), max_features=1000)

# Transforming our 'tokens' column into a TF-IDF matrix and then a data frame
tfidf_df = pd.DataFrame(tfidf.fit_transform(tsel['text_result']).toarray(), 
                        columns=tfidf.get_feature_names())
```

```python
tfidf_df
```

```python
tfidf_vectors = tfidf.fit_transform(tsel['text_result'])
```

```python
vect_score = np.asarray(tfidf_vectors.mean(axis=0)).ravel().tolist()
tsel_weight = pd.DataFrame({'term': tfidf.get_feature_names(), 'weight': vect_score})
tsel_weight.sort_values(by='weight',ascending=False,inplace=True)
tsel_weight.reset_index(inplace=True)
tsel_weight.drop(['index'], axis='columns', inplace=True)
tsel_weight.head()
```

```python
tfidf_vec = tfidf_df.copy()
```

```python
tsel['Sinyal'].value_counts()
```

```python
X = tfidf_vec
y = tsel['Sinyal']
```

```python
tsel.iloc[3577]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=25)
```

## Klasifikasi MNB

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()

# params_mnb = {'alpha': [0.1],
#               'fit_prior': [True, False]}
```

```python
# from sklearn.model_selection import GridSearchCV
# clf = GridSearchCV(model, params_mnb, cv=10)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_true = y_test
```

### random 25

```python
#example training
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
#[None, 'micro', 'macro', 'weighted'].
print(confusion_matrix(y_true,y_pred))
print(classification_report(y_true,y_pred))
print("Precision: ",precision_score(y_true, y_pred, average="macro"))
print("Recall: ",recall_score(y_true, y_pred, average="macro"))
print("F1: ",f1_score(y_true, y_pred,  average="macro"))
print("Accuracy: ",accuracy_score(y_true, y_pred))
```

```python
#### ROC-AUC

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')
y_pred_roc = enc.fit_transform(np.array(y_pred).reshape(-1,1))
Y_pred = y_pred_roc.toarray()

y_true_r = enc.fit_transform(np.array(y_true).reshape(-1,1))
y_true_roc = y_true_r.toarray()

from sklearn.metrics import roc_auc_score
roc_auc_score(y_true_roc,Y_pred, multi_class='ovr')
```
